```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df=pd.read_csv('Iris 2.csv')
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,0:4].values
y=df.iloc[:,-1].values
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
le=LabelEncoder()
y=le.fit_transform(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
lda.fit_transform(x_train,y_train)
y_pred=lda.transform(x_test)
y_pred
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_train,cmap='rainbow',alpha=0.7,edgecolor='b')
plt.title('train')
plt.show()
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,cmap='rainbow',alpha=0.7,edgecolor='b')
plt.title('test')
plt.show()
```

```python
acc_train=lda.score(x_train,y_train)
acc_train
```

```python
acc_test=lda.score(x_test,y_test)
acc_test
```

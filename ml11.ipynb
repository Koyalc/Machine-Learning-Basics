```python
import numpy as np
R=np.array([[-1,-1,-1,-1,0,-1],
           [-1,-1,-1,0,-1,100],
           [-1,-1,-1,0,-1,-1],
           [-1,0,0,-1,0,-1],
           [0,-1,-1,0,-1,100],
           [-1,0,-1,-1,0,100]])
Q=np.zeros_like(R,dtype="float")
gamma=0.8

for _ in range(10000):
    state=np.random.randint(0,Q.shape[0])
    actions=np.where(R[state]>=0)[0]
    action=np.random.choice(actions)
    Q[state,action]=R[state,action]+gamma*np.max(Q[action])
print("Updated Q matrix:",Q/np.max(Q)*100)

c_state=2
path=[c_state]
while c_state!=5:
    c_state=np.argmax(Q[c_state])
    path.append(c_state)
print("path to follow:",path)
 
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score 
df = pd.read_csv("Heart.csv") 
df.head() 

```

```python
df_encoded=df.copy() 
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() 
print(df_encoded.columns) 
```

```python
df_encoded['Age']=le.fit_transform(df_encoded['Age'].values) 
df_encoded['Sex']=le.fit_transform(df_encoded['Sex'].values) 
df_encoded['ChestPain']=le.fit_transform(df_encoded['ChestPain'].values) 
df_encoded['RestBP']=le.fit_transform(df_encoded['RestBP'].values) 
df_encoded['Chol']=le.fit_transform(df_encoded['Chol'].values) 
df_encoded['Fbs']=le.fit_transform(df_encoded['Fbs'].values) 
df_encoded['RestECG']=le.fit_transform(df_encoded['RestECG'].values) 
df_encoded['MaxHR']=le.fit_transform(df_encoded['MaxHR'].values) 
df_encoded['ExAng']=le.fit_transform(df_encoded['ExAng'].values) 
df_encoded['Oldpeak']=le.fit_transform(df_encoded['Oldpeak'].values) 
df_encoded['Slope']=le.fit_transform(df_encoded['Slope'].values) 
df_encoded['Ca']=le.fit_transform(df_encoded['Ca'].values) 
df_encoded['Thal']=le.fit_transform(df_encoded['Thal'].values) 
df_encoded['AHD']=le.fit_transform(df_encoded['AHD'].values) 
df_encoded.head()
```

```python

X=df_encoded.iloc[:,1:-1] 
y=df_encoded.iloc[:,-1] 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0) 
rf=RandomForestClassifier(oob_score=True) 
rf.fit(X_train,y_train) 
rf.oob_score_
```

```python
y_pred=rf.predict(X_test) 
accuracy_score(y_test,y_pred) 
```
